# Preparing Batsmens Data

```python
batters=df.groupby('batter')
```

```python
batter=batters.sum()['batsman_run'].sort_values(ascending=False)
batter=batter.reset_index()
batter=batter
batter
```

```python
sixes1=df[(df['batsman_run']==6)] 
sixes=sixes1.groupby('batter')['batsman_run'].count().sort_values(ascending=False).reset_index()
sixes=sixes.rename(columns={'batsman_run':'Total_number_of_sixes'})
sixes
```

```python
Four1=df[(df['batsman_run']==4)] 
Four=Four1.groupby('batter')['batsman_run'].count().sort_values(ascending=False).reset_index()
Four=Four.rename(columns={'batsman_run':'Total_number_of_Four'})
Four
```

```python
top20=batter.merge(sixes,on='batter').merge(Four,on='batter')
```

```python

batsmens=df['batter'].unique()
```

```python
lst = []
def player(name):
    temp_df=df[df['batter']==name]
    highest_score=temp_df.groupby('ID')['batsman_run'].sum().sort_values(ascending=False).head(1).values[0]
    return highest_score
for name in batsmens:
    lst.append([name,player(name)])
```

```python
highest_score=pd.DataFrame(lst,columns=['batter','Highest_score'])
highest_score=highest_score.sort_values('Highest_score',ascending=False)
highest_score=highest_score.reset_index()
highest_score=highest_score.drop(columns='index')
highest_score
```

```python
# droping rows where deliveries was wide 
df['extra_type'].unique()
```

```python
batsmens=df['batter'].unique()
extra=df[(df['extra_type']=='wides')]
df1=df.drop(extra.index).reset_index()
df1=df1.drop(columns='index')
```

```python
df1
```

```python
df['kind'].unique()
```

```python
#Each Match Performance
```

```python
lst = []

for name in batsmens:
    player = df1[df1['batter'] == name]
    runs_by_match = player.groupby('ID')
    bat_by_match=runs_by_match.agg({'batsman_run':'sum','ballnumber':'count'}).reset_index()
    
    # Append the player's name and the runs for each match to the list
    for index, row in bat_by_match.iterrows():
        lst.append([name, row['ID'],row['batsman_run'],row['ballnumber']])
```

```python
Batsmens=pd.DataFrame(lst,columns=['batter','ID','Runs_score','Number_of_balls'])
Batsmens['MatchNumber'] = Batsmens.groupby('batter').cumcount() + 1
```

```python

Batsmens
```

```python
type(Batsmens)
```

```python
Batsmens['50s']=np.where(Batsmens['Runs_score']>=50,1,0)
Batsmens['100s']=np.where(Batsmens['Runs_score']>=100,1,0)
```

```python
Batsmens['30s']=np.where((Batsmens['Runs_score']>30) & (Batsmens['Runs_score']<50) ,1,0)
Milestones=Batsmens.groupby('batter')
Milestones=Milestones.agg({'50s':'sum','100s':'sum','30s':'sum','MatchNumber':'last'})
```

```python
lst = []

for name in batsmens:
    player = df1[df1['player_out'] == name]
    out_by_match = player.groupby('batter')['isWicketDelivery'].count().sum()
    lst.append([name,out_by_match])
```

```python
number_of_outs=pd.DataFrame(lst,columns=['batter','Number_of_outs'])
number_of_outs
```

```python
stats=Batsmens.groupby('batter')['Number_of_balls'].sum().reset_index().sort_values('Number_of_balls',ascending=False)
stats
```

```python
allbat=batter.merge(sixes,on='batter').merge(Four,on='batter').merge(highest_score,on='batter').merge(stats,on='batter').merge(number_of_outs,on='batter').merge(Milestones,on='batter')
allbats=allbat.rename(columns={'Number_of_outs':"Number_of_times_dismissed"})
allbats
```

```python
allbats['Average']=np.round(allbats['batsman_run']/allbats['Number_of_times_dismissed'],2)
allbats['Strike_rate']=np.round(allbats['batsman_run']/allbats['Number_of_balls']*100,2)
allbats.to_excel(r'C:\Users\HP\OneDrive\Desktop\Resume Project\IPL\BatsmenData.xlsx')
allbats
```

# Preparing Bowlers Data

```python
bowler=df.groupby('bowler')['isWicketDelivery']
bowler=bowler.sum().sort_values(ascending=False)
bowler=bowler.reset_index()
bowler
```

```python
bowler_data=df[(df['kind']=='retired hurt')|(df['kind']=='run out')|(df['kind']=='retired out')|(df['kind']=='obstructing the field')]
bowler_data=df.drop(bowler_data.index)
bowler_data
```

```python
top_20_bowlers=bowler_data.groupby('bowler')['isWicketDelivery'].sum().reset_index().sort_values('isWicketDelivery',ascending=False)
```

```python
df['extra_type']
```

```python
bowler_data=df[(df['extra_type']=='penalty')|(df['extra_type']=='byes')|(df['extra_type']=='legbyes')]
df2=df.drop(bowler_data.index)
df2
```

```python
bowler=df['bowler'].unique()
```

```python
lst = []

for name in bowler:
    player = df2[df2['bowler'] == name]
    data_by_match = player.groupby('ID')
    data_by_match=data_by_match.agg({'total_run':'sum'}).reset_index()
    # Append the player's name and the runs for each match to the list
    for index, row in data_by_match.iterrows():
        lst.append([name, row['ID'],row['total_run']])
```

```python
Bowler=pd.DataFrame(lst,columns=['bowler','Match_id','Runs'])
Bowler.sort_values('bowler',ascending=True)
```

```python
dot_ball=df2[df2['total_run']==0]
dot_balls=dot_ball.groupby('bowler')['total_run'].count().reset_index()
dot_balls=dot_balls.rename(columns={'total_run':'DotBalls'})
dot_balls
```

```python
len(bowler)
```

```python
bowler_data=df[(df['extra_type']=='noballs')|(df['kind']=='run out')]
df2=df.drop(bowler_data.index)
df2
```

```python
lst = []

for name in bowler:
    player = df2[df2['bowler'] == name]
    data_by_match = player.groupby('ID')
    data_by_match=data_by_match.agg({'ballnumber':'count','isWicketDelivery':'sum'}).reset_index()
    # Append the player's name and the runs for each match to the list
    for index, row in data_by_match.iterrows():
        lst.append([name, row['ID'],row['ballnumber'],row['isWicketDelivery']])
```

```python
Bowler1=pd.DataFrame(lst,columns=['bowler','Match_id','Numberofballs','no_of_wicket'])
Bowler1.sort_values('bowler',ascending=True)
```

```python
Bowler1['Numberofballs'].max()
```

```python
Bowler1['Numberofballs']=Bowler1['Numberofballs'].where(Bowler1['Numberofballs']<=24,24)
Bowler1['3+ Wickets']=np.where((Bowler1['no_of_wicket']>=3) &(Bowler1['no_of_wicket']<5),1,0)
Bowler1['5+ Wickets']=np.where(Bowler1['no_of_wicket']>=5,1,0)
Bowler1['MatchNumber']=Bowler1.groupby('bowler').cumcount()+1
```

```python
Bowler_data=Bowler.merge(Bowler1,how='inner',on=['bowler','Match_id'])
Bowler_data['Overs']=np.round(Bowler1['Numberofballs']/6,2)
Bowler_data['Overs']=Bowler_data['Overs'].apply(lambda x: math.floor(x * 10) / 10)
```

```python
Bowler_data['Econ']=np.round(Bowler_data['Runs']/Bowler_data['Overs'],2)
Bowler_data
```

```python
Bowler_data['Performance'] = Bowler_data['no_of_wicket'].astype(str) + '/' + Bowler_data['Runs'].astype(str)
```

```python
Bowler2=Bowler_data.groupby('bowler')
Bowler_data1=Bowler2.agg({
    'Runs': 'sum',
    'Numberofballs': 'sum',
    'no_of_wicket': 'sum',
    '3+ Wickets': 'sum',
    '5+ Wickets': 'sum',
    'Overs': 'sum',
    'Econ': 'mean',
    'MatchNumber':'count',
    'Performance':'max'
})
```

```python
Bowler_data1=Bowler_data1.reset_index().sort_values('no_of_wicket',ascending=False)
Bowler_data1=Bowler_data1.drop(columns='Overs')
```

```python
Bowler_data1=Bowler_data1.reset_index()
Bowler_data1=Bowler_data1.drop(columns='index')
```

```python
dot_balls
```

```python
Bowler_data1=pd.merge(Bowler_data1,dot_balls,on='bowler',how='left')
Bowler_data1
```

```python
Bowler_data1['Econ']=np.round(Bowler_data1['Econ'],2)
Bowler_data1['Strike_rate']=np.round(Bowler_data1['Numberofballs']/Bowler_data1['no_of_wicket'],2)
Bowler_data1['Average']=np.round(Bowler_data1['Runs']/Bowler_data1['no_of_wicket'],2)
Bowler_data1['DotBalls']=Bowler_data1['DotBalls']
```
